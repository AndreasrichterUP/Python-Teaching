# Bsp einer Matrix zerlegung in Dreiecks-

```python
# Singular Value Decomp Example
#===============================================
#http://stackoverflow.com/questions/1730600/principal-component-analysis-in-python
#http://sebastianraschka.com/Articles/2014_pca_step_by_step.html


Im=skio.imread('feynman.png')
print Im.shape

U,s,V = lina.svd(Im)


plt.figure(figsize=(20,10))
plt.imshow(Im, cmap = 'gray')
plt.title('feynman')
plt.show()
print 'V:', V.shape[0]

# Full image 
S=np.zeros(Im.shape);
#S[0:V.shape[0],:] = np.diag(s)

# some eigenvalues
n=25;  
for i in range(1,n):
    S[i,i]=s[i]

print U.shape
print s.shape
print V.shape
print S.shape

#A= np.mat(U)* np.mat(S)
#A.shape
#BildPCA=np.uint8(np.mat(U)* np.mat(S)*np.mat(V))

BildPCA=np.mat(U)* np.mat(S)*np.mat(V)      
print 'Shape of Output:{}  Type of Output:{}   Type Of Output (with bytescale):{}'.format(BildPCA.shape, BildPCA.dtype, bytescale(BildPCA).dtype)
plt.figure(figsize=(20,10))
#plt.imshow(bytescale(BildPCA),cmap='gray')
plt.imshow(BildPCA,cmap='gray')

plt.title('feynman')

plt.show()
#===============================================
```
