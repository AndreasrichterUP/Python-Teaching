# Singularvalue Decomposition of an image matrix

Images can be handled in form of numpy array's (big matrix) so every algebraic matrix operation is possible to manipulate an images. The above imported image is now available in form of a 620 rows by 475 cols numpy array.

To find out what is meant by the topic **Singular-Value-Decomposition** of a matrix I have to explain first the terms **Eigenvalues** and **Singularvalues**. Both are closely connected.

The so called **Eigenvalues** $\lambda$ of a squared matrix $A_{n,n}$ are characteristic numbers for every matrix or array that fullfill the following equation:  

$A_{n,n} \vec x = \lambda \vec x$

Now I am interested in the values of $\lambda$ and $\vec x$ that fulfill this equation.
To answere this question I have to find at first one of the solutions $\lambda_i$ and with this I can find the corresponding $ \vec x_i$. There are as many solutions $\lambda_i$ as there are row/cols in the matrix $A_{n,n}$, in this case $n$.

To solve this equation according to $\lambda$ the following manipulations can be done:

$A_{n,n} \vec x = \lambda E_{n,n} \vec x$  ......(where $E_{n,n}$ is the Unit-Matrix)

$A_{n,n} \vec x - \lambda E_{n,n} \vec x= 0$  

$(A_{n,n} - \lambda E_{n,n} ) \cdot \vec x = 0$  

accordingly whether $\vec x $ or $(A_{n,n} - \lambda E_{n,n} )$ have to be null.

Ignoring the trivail solution of $\vec x =0$ makes sense because it doesn't give any new information.

$(A_{n,n} - \lambda E_{n,n} ) = 0$ is a far more promising. 

This can be done by evaluating the determinat of the algebraic equation which leads to the so called **characteristic polynom** and is expressed for every symmetric matrix **$A_{n,n}$** alike this:
                $ p_n (\lambda) = det(A_{n,n} - \lambda \cdot E_{n,n} )$

==> **characteristic polynom** : $ p_n (\lambda) = det(A_{n,n} - \lambda \cdot E_{n,n} )$

For example the determinant det(...) of the matrix $A = \begin{pmatrix}
a & b\\
c & d
\end{pmatrix}$  is det (A) = $\begin{vmatrix}
a & b\\
c & d
\end{vmatrix}$ = $ a \cdot d - c \cdot b$

for further information on higher dimensions determinants see also: https://de.wikipedia.org/wiki/Determinante 

To calculate the Eigenvalues, we need the characteristic polynom to be zero, so:                 $det(A_{n,n} - \lambda \cdot E_{n,n} ) = 0$
                
This equation results in a **polynom in $\lambda$** of **degree n** according to the dimensions of $A_{n,n}$

Polynoms of **degree n** always have **n - Null crossings **  $\lambda_i $ which are the **Eigenvalues** I am looking for.

When a found solution $\lambda_i$ is placed in the original equation from above the so called Eigenvector $\vec x_i$ can be determined.

The Eigenvalues can be sorted according to their size in assumption that the biggest one represents the most information. The corresponding Eigenvectors are sorted in the same order designing a new matrix. This matrix of Eigenvectors is the transformation matrix of the Principal Component Analysis (PCA - See Notebook **-07a-PCA-Example**)

Import of modules and libraries...

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#%matplotlib notebook

from skimage import io as skio
from scipy import linalg as lina

from scipy.misc import bytescale
```

Reading an image to a numpy array 

```python
Im=skio.imread('figures/feynman.png')
print(Im.shape)
print(Im.dtype)
plt.imshow(Im, cmap='gray')
plt.show()
```

Here is an example how to calculate the Eigenvalues and the Eigenvectors of a small matrix $A_{3,3}$. So three Eigenvalues and three Eigenvectors have to be determined.

```python
A33=np.array([[0,2,-1],[2,-1,1],[2,-1,3]])
```

To check my results I can use the scipy.linalg function **scipy.linalg.eigvals()**

```python
#print(lina.eigvals(A33))
print(lina.eig(A33)[1])
print(25*'=')
print(lina.eigvals(A33))
print(25*'=')
print(lina.eigvals_banded(A33))
```

So if I am interested to do the same with a real image which is normally not squared but rectangular, I have to concider to calculate the Eigenvector equations once from the left side but also from the right side (normal approach - in case of a squared matrix). 

This will result in two sets of Eigenvectors the **left-sided** and the **right-sided Eigenvectors**. 

The Eigenvalues itself 

**Matrix decomposition into three parts (left-Eigenvalues, Main-diagonal-singularvalues, right-Eigenvalues)** 

U... left Eigenvectors

s... singular Values

V... right Eigenvalues  

With these the decomposition is done with a function from the scipy packedge **linalg** for linear algebraic manipulations  


```python
# Singular Value Decomp Example
#===============================================
#http://stackoverflow.com/questions/1730600/principal-component-analysis-in-python
#http://sebastianraschka.com/Articles/2014_pca_step_by_step.html


U,s,V = lina.svd(Im)
```

```python
print(U[0:5,0:5])

print(U.shape)
```

```python
print(s[0:10])

print(s.shape)
```

```python
print(V[0:5,0:5])

print(V.shape)
```

So as I can see from the dimensions **U-Mat(620x620)**, the **s-Vec(475,1)** and the **V-Mat(475,475)** are the results of the decomposition.

```python
plt.figure(figsize=(20,10))
plt.imshow(Im, cmap = 'gray')
plt.title('feynman')
plt.show()
print( 'V:', V.shape[0])

# Full image 
S=np.zeros(Im.shape);
#S[0:V.shape[0],:] = np.diag(s)

# some eigenvalues
n=2;  
for i in range(1,n):
    S[i,i]=s[i]

print( U.shape)
print( s.shape)
print( V.shape)
print( S.shape)

#A= np.mat(U)* np.mat(S)
#A.shape
#BildPCA=np.uint8(np.mat(U)* np.mat(S)*np.mat(V))

#BildPCA=np.mat(U)* np.mat(S)* np.mat(V)
BildPCA=np.dot(np.dot(U,S),V)

print( 'Shape of Output:{}  Type of Output:{}   Type Of Output (with bytescale):{}'.format(BildPCA.shape, BildPCA.dtype, bytescale(BildPCA).dtype))
plt.figure(figsize=(20,10))
#plt.imshow(bytescale(BildPCA),cmap='gray')
plt.imshow(BildPCA,cmap='gray')

plt.title('feynman')

plt.show()
#===============================================
```
