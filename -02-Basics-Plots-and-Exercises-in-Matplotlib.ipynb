## Plotting of 2D-Data in Singleplots <a name="2dplots" ></a>

matplotlib is imported by the lines below

```python
import matplotlib.pyplot as plt
```

where **plt** is used as an abreviation to save time in writing the source code. To make sure a diagram, an image or a plot is shown directly in the notebook below the code, the 'inline' parameter must be set: 

```python
%matplotlib inline
```

In the following example random numbers are generated, assigend to a linear function and added with random gaussian noise

```python
import numpy as np
```

The function **'plt.plot(x,y,'.b')'** is used with the paraters **'x:x-values'** at first and than **'y:y-values'** secondly assigned to the corresponding x and at third place the symbol used to display the datapoints in the plot, here a '.' means a small dot and '.b' specifies a small dot of color blue. 

```python
x=10*np.random.rand(20,1)   # uniform distributet samples between [0 ... 10]
y=6.3*x+15
plt.plot(x,y,'b.')
plt.show()
```

The plot shows a perfectly alligned line. So now I will add some confusion to the y data in form of small random variations and repeat the plot. Everytime I run the cell **'y'** will be updated.

```python
y1=y+5*np.random.randn(20,1) # add random gaussian noise as random data errors
plt.plot(x,y1,'b.')
plt.show()
```

## Plots arranged side-by-side or in a figure <a name="figure" ></a>

if more than one plot is needed the functions  **figure()** and **subplot()** is needed

The numbering of the "subplot(221)" means a 2 by 2 grid (22x-> first two numbers and the index x of current subplot [1...4] )

More than one plot can be arranged by the use of the subplot routine. 

```python
%matplotlib inline

x=10*np.random.rand(20,1)   # uniform distributet samples between [0...10]
y=6.3*x+15
y=y+10*np.random.randn(20,1)
x=np.ravel(x)
m,b=np.polyfit(x,y,1)
plt.plot(x,y,'ro')           # single plot 
plt.title('Random Points in single-plot')

plt.figure()   # figure with subplots
plt.subplot(221),plt.plot(x,y,'ro')
plt.title('Points')
plt.subplot(222),plt.plot(x,m*x+b,'b')
plt.title('Regression-line')
plt.subplot(223),plt.plot(x,y,'ro'),plt.plot(x,m*x+b,'b')
plt.title('Points and Regression-line')

#plt.show()
```

## Changing the size of a figure <a name="figuresize" ></a>

```python
plt.figure(figsize=(16,12))   # figure with subplots
plt.subplot(221),plt.plot(x,y,'ro')
plt.title('Points')
plt.subplot(222),plt.plot(x,m*x+b,'b')
plt.title('Regression-line')
plt.subplot(223),plt.plot(x,y,'ro'),plt.plot(x,m*x+b,'b')
plt.title('Points and Regression-line')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scripts.plot_2d_separator import plot_2d_separator
n=200
# normal distribution and ring around it
class_3 = 0.6 * np.random.randn(n,2)
r = 0.8 * np.random.randn(n,1) + 5
angle = 2*np.pi * np.random.randn(n,1)
class_4 = np.hstack((r*np.cos(angle),r*np.sin(angle))) 
labels_b = np.hstack((np.ones(n),-np.ones(n)))

plt.figure(figsize=(10,5))
plt.plot(class_3[:,0],class_3[:,1],'r.')
plt.plot(class_4[:,0],class_4[:,1],'b.')
plt.xlim(-10,10)
plt.show()
```

```python
'''
======================
Triangular 3D surfaces
======================

Plot a 3D surface with a triangular mesh.
'''

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

n_radii = 8
n_angles = 36

# Make radii and angles spaces (radius r=0 omitted to eliminate duplication).
radii = np.linspace(0.125, 1.0, n_radii)
angles = np.linspace(0, 2*np.pi, n_angles, endpoint=False)

# Repeat all angles for each radius.
angles = np.repeat(angles[..., np.newaxis], n_radii, axis=1)

# Convert polar (radii, angles) coords to cartesian (x, y) coords.
# (0, 0) is manually added at this stage,  so there will be no duplicate
# points in the (x, y) plane.
x = np.append(0, (radii*np.cos(angles)).flatten())
y = np.append(0, (radii*np.sin(angles)).flatten())

# Compute z to make the pringle surface.
z = np.sin(-x*y)

fig = plt.figure(figsize=(10,10))
ax = fig.gca(projection='3d')

ax.plot_trisurf(x, y, z, linewidth=0.2, antialiased=True)

plt.show()
```

## Interactive Plots <a name="interactiveplots" ></a>

```python
%matplotlib widget
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
line, = ax.plot(x, np.sin(x))

def update(w = 1.0):
    line.set_ydata(np.sin(w * x))
    fig.canvas.draw_idle()

interact(update);
```
